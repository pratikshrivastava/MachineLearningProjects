Check for the correlation between the different features

The AvgCarelessness is highly correlated to the Aveknoweldge. 
It seems the students with high knowedge of the concepts tends to make more careless mistakes. 
The student with knowledge of the topic also have higher probability of getting the the correct answers. 

The students with high avgres also tends to have -ve effect for the off task. 
so as the students who are easily bored also have -ve effect for the Gaming of the system. 

```python
f, ax = plt.subplots(figsize=(10, 8))
corr = xDF.corr()
sns.heatmap(corr,cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
```

Generated pairplot for the different features for checking the relationshup among the features.

```python
sns.pairplot(xDF,diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```
